```python
#Fig 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import wofz

# Define a range of values for t (time) spanning 24 hours
start_time = 0.0
end_time = 24 * 3600  # 24 hours in seconds
num_points = 100  # You can adjust the number of points
t_values = np.linspace(start_time, end_time, num_points)

# Define a list of different b values, excluding r_b=1
b_values = [10, 100, 1000, 10000, 1000000000]

# Set k_d to 1
k_d = 1

# Create a color map for plotting
colors = plt.cm.viridis(np.linspace(0, 1, len(b_values)))

# Plot S(t) for each b value
plt.figure(figsize=(10, 6))
for i, b in enumerate(b_values):
    S_values = []
    for t in t_values:
        a1 = -b / 2 + np.sqrt(b**2 / 4 - 1)
        a2 = -b / 2 - np.sqrt(b**2 / 4 - 1)
        tau = k_d * t
        S_t = (a2 * wofz(-1j * a1 * np.sqrt(tau)) - a1 * wofz(-1j * a2 * np.sqrt(tau))) / (a2 - a1)
        S_values.append(np.real(S_t) * 100)  # Take the real part and convert to percentage

    S_values = np.array(S_values)

    # Plot S(t) against t with a label indicating the b value
    label = f'$r_b$ = {b}' if b != float('inf') else '$r_b$ = infinity'
    plt.plot(t_values / 3600, np.real(S_values), label=label, color=colors[i], linewidth=2)

plt.xlabel('Time (hours)', fontsize=18)
plt.ylabel('Capture probability (%)', fontsize=18)
plt.xticks(fontsize=14)  # Adjust x-axis tick font size
plt.yticks(fontsize=14)  # Adjust y-axis tick font size
plt.ylim(-5, 105)  # Set the y-axis limits to -5% to 105%
plt.grid(True)
plt.legend(fontsize=16)
plt.savefig('S_vs_t_different_rb.png')
plt.show()
```

```python
#Fig 3
import math
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas for working with data


# Define parameters
n = 16
m = 5
d_target = 25  # nm
K_1 = 1e-3  # M
a_0 = 0.4  # nm
omega_LU = 0.03
alpha = 1

def K_n(K_1, small_omega_lu, C_n, m, n):
    "equation s11"
    return ((K_1**n)* (small_omega_lu)**n)/(C_n * m**n)

def calculate_Cn(n, d_inter, d_target, a0, r_ete):
    term1 = n * (2**(3/2)) / (8 * math.pi**3 * n**2)
    term2 = ((2 * math.pi) / 3)**2
    term3 = (d_inter / d_target)**(-2)
    term4 = 0.5 * n**(-3/2) * (3 / (2 * math.pi))**(3/2 * (n - 1))
    term5 = (2**7 / (3 * math.pi**2 * n))**(n / 2)
    term6 = 1 / math.sqrt(math.pi) * gamma((1 + n) / 2)
    term7 = (a0 / d_target)**n
    term8 = d_target**(-3 * (n - 1))
    term9 = (d_inter / d_target - 1)**(7 - 4 * n)
    term10 = ((d_inter - d_target) / r_ete)**(4 * n - 7)
    term11 = math.exp(-1.5 * n * ((d_inter - d_target) / r_ete)**2)

    result = term1 * term2 * term3 * term4 * term5 * term6 * term7 * term8 * term9 * term10 * term11

    return result

# Define d_inter values
d_inter = np.linspace(0.5, 50, 10000)

# Calculate r_ete and K_values
n_values = [16]
K_values = []

for n in n_values:
    r_ete = np.sqrt(3 * n / (4 * n - 7)) * np.abs(d_inter - d_target)
    C_n = [calculate_Cn(n, d_inter[i], d_target, a_0, r_ete[i]) for i in range(len(d_inter))]
    K_n_values = [K_n(K_1, omega_LU, C_n[i], m, n) for i in range(len(d_inter))]
    K_values.append(K_n_values)

# Find the index of the closest d_inter value to 20 nm
index_d_inter_20 = np.argmin(np.abs(d_inter - 20))

# Get the corresponding K_n value
K_n_at_d_inter_20 = K_values[0][index_d_inter_20]

# Create the plot
plt.figure(figsize=(10, 6))
for i, n in enumerate(n_values):
    plt.plot(d_inter, K_values[i], label=f'n={n}', linewidth=3, color='#F88088')

# Add a vertical dashed line at x = 20 (d_inter = 20 nm)
plt.axvline(x=20, color=(141/255, 194/255, 0/255), linewidth=2, linestyle='--', label=r'$d_{inter} = 20$ nm')

# Display the intersection point on the plot
label = r'$K_{n}$ = $%.2e$ M' % K_n_at_d_inter_20
plt.axhline(y=K_n_at_d_inter_20, color=(160/255, 221/255, 255/255), linewidth=2, linestyle='--', label=label)


plt.xlabel(r"Inter-binder radius $d_{inter}$ (nm)", fontsize=16)
plt.ylabel(r"$K_{n}$ (M)", fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(1e-38, 1e-1)
plt.yscale('log')
plt.axhline(y=1e-3, color='#008080', linestyle='--', label=r'$K_1$ = $10^{-3}$ M', linewidth=2)
plt.legend(fontsize=16)
plt.savefig('plot_of_K_n.png')
plt.show()
```
